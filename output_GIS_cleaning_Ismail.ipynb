```python
# Import packages
import pandas as pd
import datetime
```

```python
# Import dataset
data = pd.read_excel('output_GIS.xlsx')
```

```python
# Explore data
data
```

```python
def landuse_split(data):

    item2_list = []

    item_number = 0

    for item in data['Landuse']:
        item_number += 1
        item_split = item.split(';')
        for item2 in item_split:
            item2_dict = {'Item_Number': item_number, 
                          'Item2': item2}
            item2_list.append(item2_dict)

    item2_df = pd.DataFrame(item2_list)
    
    return item2_df
```

```python
def landuse_split_values(data):
    
    item2_df = data
    item_number = 0
    item_list1 = []
    item_list2 = []
    item_list3 = []

    for item in item2_df['Item2']:
        item_number += 1
        item_split = item.split(',')
        item_list1.append(item_split[0])
        item_list2.append(item_split[1])
        item_list3.append(item_split[2])
    
    item2_df['LandType'] = item_list1
    item2_df['LandValue1'] = item_list2
    item2_df['LandValue2'] = item_list3
    
    return item2_df
```

```python
def landuse_merge(original_data, item2_df):

    data_merge = pd.merge(original_data, item2_df, how = 'left', left_on = ['OBJECTID'], right_on = ['Item_Number']).reset_index(drop = True)
    
    return data_merge
```

```python
item_df = landuse_split(data)
item2_df = landuse_split_values(item_df)
data_merge = landuse_merge(data, item2_df)
```

```python
# Explore output data
data_merge
```

```python
# Export as csv
data_merge.to_csv('output_GIS_' + datetime.datetime.now().strftime('%Y%m%d%H%M%S') + '.csv', index = False)
```
